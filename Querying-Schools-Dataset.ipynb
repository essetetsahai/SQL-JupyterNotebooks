Query database system catalog to retrieve table metadata

```python
#Query list of all tables in the db2 schema
%sql select TABSCHEMA, TABNAME, CREATE_TIME from SYSCAT.TABLES where TABSCHEMA = 'NTL45502'
```

```python
#Or retrieve list of tables where schema is not created by the system:
%sql select TABSCHEMA, TABNAME, CREATE_TIME from SYSCAT.TABLES \
    where TABSCHEMA not in ('SYSIBM', 'SYSCAT', 'SYSSTAT', 'SYSIBMADM', 'SYSTOOLS', 'SYSPUBLIC')
```

```python
#Or query the specific table to verify exists in the database
%sql select * from SYSCAT.TABLES where TABNAME = 'SCHOOLS'
```

Query database system catalog to retrieve column metadata

```python
#How many columns does the SCHOOLS table have?
%sql select count(*) as NUM_OF_COLUMNS from SYSCAT.COLUMNS where TABNAME = 'SCHOOLS'
```

```python
#Retrieve list of columns with their names, types and lengths from the SCHOOLS table
%sql select COLNAME, TYPENAME, LENGTH from SYSCAT.COLUMNS where TABNAME = 'SCHOOLS'
```

```python
#How many elementary schools are in the dataset?
%sql select count(*) as NUM_OF_ES from SCHOOLS where SCHOOL_TYPE = 'ES'
```

```python
#What is the highest safety score?
%sql select MAX(Safety_Score) AS HIGHEST_SAFETY_SCORE from SCHOOLS
```

```python
#What schools have the highest safety score
%sql select Name_of_School, Safety_Score from SCHOOLS where Safety_Score = 99
```

```python
#or
%sql select Name_of_School, Safety_Score from SCHOOLS where \
    Safety_Score= (select MAX(Safety_Score) from SCHOOLS)
```

```python
#Which 10 schools have the highest Average_student_attendance
%sql select Name_of_School, Average_Student_Attendance from SCHOOLS \
    order by Average_Student_Attendance desc nulls last limit 10
```

```python
#Which 5 schools have the lowest average_student_attendance (sort in ascending order)

%sql select Name_of_School, Average_Student_Attendance from SCHOOLS \
    order by Average_Student_Attendance \
    fetch first 5 rows only
```

```python
#Remove the '%' sign from the above result
%sql select Name_of_School, REPLACE(Average_Student_Attendance, '%', '') as AVG_ATND \
    from SCHOOLS \
    order by Average_Student_Attendance \
    fetch first 5 rows only
```

```python
#Which schools have avg student attendance < 70%?
#Cast Avg student attendance as double
%sql select Name_of_School, Average_Student_Attendance \
    from SCHOOLS \
    where CAST(REPLACE(Average_Student_Attendance, '%', '') AS DOUBLE) < 70 \
    order by Average_Student_Attendance
```

```python
%sql select Name_of_School, Average_Student_Attendance \
    from SCHOOLS \
    where DECIMAL(REPLACE(Average_Student_Attendance, '%', '')) <70 \
    order by Average_Student_Attendance
```

```python
#What is the total college enrollment of each community area?
%sql select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT \
    from SCHOOLS \
    group by Community_Area_Name
```

```python
#Get the 5 community areas with the lowest college enrollments in ascending order
%sql select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT \
    from SCHOOLS \
    group by Community_Area_Name \
    order by TOTAL_ENROLLMENT \
    fetch first 5 rows only
```

```python
#List the 5 schools with lowest safety scores
%sql select Name_of_School, Safety_Score \
    from SCHOOLS \
    order by Safety_Score \
    LIMIT 5
```

```python
#What is the hardship index for the community area with college enrollment  of 4368
%sql select hardship_index, ca as community_area_number \
    from chicago_socioeconomic_data CD, schools CPS \
    where CD.ca = CPS.community_area_number \
    and college_enrollment = 4368
```

```python
#What is the hardship index for the community area which has the school with the highest enrollment
%sql select ca, community_area_name, hardship_index from chicago_socioeconomic_data \
    where ca in (select community_area_number from SCHOOLS order by college_enrollment desc limit 1)
```
